```python
week_work = pd.DataFrame({'Pre-Requisite':[40, 5, 0, 0],
              'Installation':[0, 25, 5, 0],
              'Configuration':[0, 5, 30, 5],
              'Sanity Check':[0, 0, 5, 21]}, index = ['Week1', 'Week2', 'Week3', 'Week4'])
```

```python
week_work1 = week_work.reindex_axis(['Pre-Requisite', 'Installation', 'Configuration', 'Sanity Check'], axis=1)
```

```python
week_work1.plot(kind='bar' )
plt.xlabel('Efforts put per Week', fontsize=15)
plt.xticks(rotation=70)
plt.ylabel('Efforts in Hour', fontsize=15)
plt.ylim(0, 55)
##plt.legend(week_work.columns)
```

```python
Total_effort = pd.DataFrame([24, 50, 20, 40], index=['Pre-Requisite', 'Installatiion', 'Configuration', 'Sanity'])
```

```python
plt.pie(Total_effort.values, labels=Total_effort.index, autopct='%1.2f%%', explode=[0,0,0,.1], shadow=True)
plt.title('Percentage Effort for each Step', fontsize=18)
```
